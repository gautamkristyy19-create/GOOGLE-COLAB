```python
import pandas as pd

# Load dataset
url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
df = pd.read_csv(url)

# Year Range
min_year = df["year"].min()
max_year = df["year"].max()

# Number of Countries
num_countries = df["country"].nunique()

# Number of Continents
num_continents = df["continent"].nunique()

print("Year Range:", min_year, "to", max_year)
print("Number of Countries:", num_countries)
print("Number of Continents:", num_continents)
```

```python
import  matplotlib.pyplot as plt
```

```python
avg_life_exp = df.groupby('continent')['lifeExp'].mean()
avg_life_exp.plot(kind='bar', color='skyblue')
plt.title('Average Life Expectancy by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Life Expectancy')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_life = df.groupby("continent")["lifeExp"].mean()

plt.figure(figsize=(8,5))
avg_life.plot(kind="bar")

plt.title("Average Life Expectancy by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=0)

plt.show()
```

```python
# Select the 3 countries
countries = ["India", "China", "United States"]

# Filter data
filtered_df = df[df["country"].isin(countries)]

# Plot
plt.figure(figsize=(10, 5))

for country in countries:
    country_data = filtered_df[filtered_df["country"] == country]
    plt.plot(
        country_data["year"],
        country_data["lifeExp"],
        marker="o",
        label=country
    )

plt.title("Life Expectancy Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["gdpPercap"], df["lifeExp"])

plt.title("GDP per Capita vs Life Expectancy")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")

plt.grid(True)
plt.show()
```
